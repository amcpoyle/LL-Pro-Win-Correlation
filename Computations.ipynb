# The Count Method

```python
# get all  unique 450mx winners
mx450_winners_only = results450mx[results450mx['wins'] >= 1]
```

```python
unique_ll_class = list(ll['race_class'].unique())
```

```python
# source: https://stackoverflow.com/questions/3869487/how-do-i-create-a-dictionary-with-keys-from-a-list-and-values-defaulting-to-say
counts = {x:0 for x in unique_ll_class}
```

```python
for rider in mx450_winners:
    # get ll results
    ll_subset = ll[ll['rider'] == rider]
    
    for i in range(len(ll_subset)):
        race_result = ll_subset.iloc[i]
        race_class = race_result['race_class']
        race_pos = race_result['overall_result']

        counts[race_class] = counts[race_class] + (1/race_pos)
```

```python
counts
```

```python
# source: https://stackoverflow.com/questions/50194261/pandas-create-dataframe-with-only-one-row-from-dictionary-containing-lists
counts_df = pd.DataFrame({k: [v] for k, v in counts.items()})
```

```python
counts_df
```

```python
# source: https://stackoverflow.com/questions/21164910/how-do-i-delete-a-column-that-contains-only-zeros-in-pandas
counts_df = counts_df.loc[:, (counts_df != 0).any(axis=0)]
```

```python
cols = ['rider']
race_classes = list(ll['race_class'].unique())
cols = cols + race_classes
```

```python
df = pd.DataFrame(columns=cols)
```

```python
df
```

```python
unique_racerx_riders = list(racerx['rider'].unique())
```

```python
for rider in unique_racerx_riders:
    ll_subset = ll[ll['rider'] == rider]
    
    classes = list(ll_subset['race_class'].unique())
    
    best_finish_dict = {}
    
    for c in classes:
        class_subset = ll_subset[ll_subset['race_class'] == c]
        # get the min row
        best_finish = min(class_subset['overall_result'])
        best_finish_dict[c] = best_finish
        
    row_data = [rider]
    for class_ in race_classes:
        # check if the class is in the dictionary
        if class_ in best_finish_dict.keys():
            row_data.append(best_finish_dict[class_])
        else:
            row_data.append(41) # changed from None... let's see how it works
            
    df.loc[len(df)] = row_data
```

```python
# create a pro_wins column
pro_wins = []
for rider in list(df['rider']):
    subset = racerx[racerx['rider'] == rider]
    
    wins = subset[subset['position'] == 1]
    
    num_wins = len(wins)
    pro_wins.append(num_wins)
```

```python
df['pro_wins'] = pro_wins
```
